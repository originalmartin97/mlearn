```python
import sympy
import numpy as np
from matplotlib import pyplot as plt
```

```python
def func(x):
    return x*sympy.sin(2*x)
```

```python
l_arr = list()
for tx in np.linspace(-np.pi, np.pi, 100):
    l_arr.append(func(tx))
```

# Plot the function on the range of -pi to pi

```python
plt.plot(l_arr)
```

# Implemented backward method for the function

```python
def backward_method(f, x, h):
    return ( (f(x) - f(x-h)) / h )
```

# Approximated f'(pi/4)

```python
approx_result = backward_method(func, np.pi/4, 0.02)
approx_result
```

# Analytical derivative with sympy

```python
X = sympy.symbols('x')
expr = X*sympy.sin(2*X)
sympy.diff(expr, X)
```

```python
def diff(x):
    return (2*x*sympy.cos(2*x)) + (sympy.sin(2*x))
```

# Exact value of f'(pi/4)

```python
analitic_result = diff(np.pi/4)
analitic_result
```

```python
# Error
abs(approx_result - analitic_result)
```
